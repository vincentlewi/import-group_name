```python
def api_time(row):
    title = row.title_only
    year = row.year
    
    parameters = {
        "t": title,
        "y": year
    }

    response = requests.get("https://www.omdbapi.com/?apikey=ae41fb6d", params=parameters)

    poster_url = response.json()['Poster']
    plot = response.json()['Plot']
    
    return poster_url, plot
#     return pd.Series([poster_url, plot])
```

```python
import concurrent.futures

def api_time(args):
    title, year = args
    
    parameters = {
        "t": title,
        "y": year
    }

    try:
        response = requests.get("https://www.omdbapi.com/?apikey=ae41fb6d", params=parameters)

        poster_url = response.json()['Poster']
        plot = response.json()['Plot']

        return poster_url, plot
    except:
        return np.nan, np.nan

# create a ThreadPoolExecutor with 10 threads
executor = concurrent.futures.ThreadPoolExecutor(max_workers=10)

# create a list of arguments for the api_time function
args_list = [(row.title_only, row.year) for _, row in df_imdb_movies2.iterrows()]

# execute the function calls in parallel
results = list(executor.map(api_time, args_list))

# unpack the results into two separate lists
poster_urls, plots = zip(*results)

# assign the new columns to the DataFrame
df_imdb_movies2['poster_url'] = poster_urls
df_imdb_movies2['plot'] = plots
```

```python
# now process imdb_rating and num_of_rating similar to CB format
df_imdb_movies_final['num_of_rating_bins'] = pd.cut(x=df_imdb_movies_final['num_of_rating'], bins= [0, 500, 1000, 10000, 100000, 99999999999999], labels= ["super_low", "low", "medium", "high", "super_high"])

num_rating_columns = pd.get_dummies(df_imdb_movies_final.num_of_rating_bins, prefix='num_rating')
# print(num_rating_columns.head())
df_imdb_movies_final = df_imdb_movies_final.join(num_rating_columns)

df_imdb_movies_final['imdb_rating_bins'] = pd.cut(x=df_imdb_movies_final['imdb_rating'], bins= [0, 2.5, 5, 7.5, 10], labels= ["low", "medium", "high", "super_high"])

imdb_rating_columns = pd.get_dummies(df_imdb_movies_final.imdb_rating_bins, prefix='imdb_rating')
# print(imdb_rating_columns.head())
df_imdb_movies_final = df_imdb_movies_final.join(imdb_rating_columns)

df_imdb_movies_final = df_imdb_movies_final.drop(['imdb_rating_bins','num_of_rating_bins'], axis=1)
```
