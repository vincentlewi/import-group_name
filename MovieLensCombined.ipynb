```python
import pandas as pd
```

```python
df_movies = pd.read_csv('ml_original/movies.csv')
```

```python
df_movies.info()
```

```python
df_movies = df_movies.drop('genres',axis=1)
```

```python
df_movies.head(50)
```

```python
df_ratings = pd.read_csv('ml_original/ratings.csv')
df_ratings.info()
```

```python
df_ratings.head(10)
```

```python
df_ratings_sort = df_ratings.sort_values(by=["movieId"])
```

```python
df_ratings_sort.head(10)
```

```python
df_movies_ratings = pd.merge(df_movies, df_ratings_sort, on = "movieId")
df_movies_ratings.head(30)
```

```python
df_tags = pd.read_csv('ml_original/tags.csv')
df_tags.info()
```

```python
df_tags.head(10)
```

```python
df_movies_tags = pd.merge(df_movies, df_tags, on = ["movieId"])
df_movies_tags.info()
```

```python
df_movies_tags.head(10)
```
