```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../datasets/movielens_original/ratings.csv')
df_movie = pd.read_csv('../datasets/movielens_original/movies.csv')
```

```python
df_movie[df_movie.title.apply(lambda x: 'spongebob' in x.lower())].to_json(orient='records')
```

```python
df = df.merge(df_movie, how="left", on="movieId")
```

```python
df.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
# dict_movies = {i: x for i, x in enumerate(df.movieId.unique())}
# dict_users = {i: x for i, x in enumerate(df.userId.unique())}

dict_movies = {}
index = 0

for ids in sorted(df.movieId.unique()):
    dict_movies[ids] = index
    index += 1

dict_users = {}
index = 0
for ids in sorted(df.userId.unique()):
    dict_users[ids] = index
    index += 1
```

```python
import tensorflow as tf
from tensorflow.keras.models import load_model

model = load_model('saved_model.h5')
```

```python
df[df.title == 'Finding Nemo (2003)'].sort_values(by='rating', ascending=False).head(10)
```

```python
df[df.userId == 53206].sort_values(by='rating', ascending=False)
```

```python
user_id = 53206
predicted_rating = model.predict([pd.Series([user_id]*len(dict_movies)), pd.Series(dict_movies.values())])
predicted_rating = pd.DataFrame(predicted_rating)
predicted_rating
```

```python
a = pd.DataFrame(dict_movies.keys(), columns=['movieId']).merge(df_movie, how="left", on="movieId")
a
```

```python
result = pd.concat([a, predicted_rating], axis=1)
movies_watched_by_user = df[df.userId == user_id]
not_watched = result[~result.movieId.isin(movies_watched_by_user.movieId.values)]
top_10 = not_watched.sort_values(0, ascending=False)[['title', 'genres']].head(10)

top_10
```
